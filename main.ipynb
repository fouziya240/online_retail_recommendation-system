### Find Popular Item

#### Globally

```python
popular_items=df.groupby('Description') ['Quantity'].sum().sort_values(ascending=False).head(10)
print("Top 10 popular_items Globally: ")
print(popular_items)
```

#### country_wise

```python
data={
    'Country': ['United Kingdom', 'United Kingdom', 'United Kingdom', 'France', 'France'],
    'Description': ['WHITE METAL LANTERN', 'CREAM CUPID HEARTS COAT HANGER','WHITE HANGING HEART T-LIGHT HOLDER',"CHILDREN'S APRON DOLLY GIRL",'BAKING SET 9 PIECE RETROSPOT'],
    'Quantity': [10,20,30,70,50]
}

def top_items_by_Country(Country,n=5):
    Country_df=df[df['Country']=='United Kingdom']
    top_items=Country_df.groupby('Description')['Quantity'].sum(). sort_values(ascending=False).head(n)
    return top_items
print('Top items in Country:' )
print(top_items_by_Country('United Kingdom'))
```

#### Month_wise

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df['Month']=df['InvoiceDate'].dt.month
```

```python
n=int(input("enter n"))
monthwise=df.groupby(['Month','Description']) ['Quantity'].sum().reset_index()
top_monthly=monthwise.sort_values(['Month','Quantity'], ascending=[True, False]).groupby('Month').head(n)
print('Top n items per month:' )
print(top_monthly)
```

### EDA

```python
 # sales by country
country_sales=df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
country_sales[1:11].plot(kind='bar')
```

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar', title='Top 10 Customers by Revenue')
```

```python
# Top 10 excluding united kingdom
plt.title("Top 10 countries by revenue(excluding UK)")
plt.ylabel("Total Revenue")
plt.xlabel("country")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Number of transaction per country

invoice_counts=df.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False)
print("\n Top 5 countries by invoice count:")
print(invoice_counts.head())
```

```python
# Returns Analysis
df['IsReturn']=df['InvoiceNo'].astype(str).str.startswith('C')
 
print("\n Number of Returns:", df['IsReturn'].sum()) 
```

```python
# monthly sales 
df.set_index("InvoiceDate",inplace=True)
monthly_sales=df.resample("M")["TotalPrice"].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot()
plt.title("Monthly sales ")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Monthly Revenue

monthly_revenue=df.groupby('InvoiceNo')['TotalPrice'].sum()

monthly_revenue.plot(figsize=(12,5))
plt.title("Monthly Revenue")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.grid(True)
plt.show()
```

```python
# top 10 most popular product
top_products=df["Description"].value_counts().head(10)
print("Top 10 products:\n",top_products)
```

```python
# plotting top products

top_products=df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_products.plot(kind='pie')
plt.title("Top 10 most sold products")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=75)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# number of unique Customers

print("\n number of unique customer:", df['CustomerID'].nunique())
```

```python
# Average items per order

basket_size=df.groupby('InvoiceNo')['Quantity'].sum().mean()

print("\n Average basket size(items per invoice):", round(basket_size,2))
```

```python
# Revenue per prooduct
product_revenue=df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
product_revenue.plot(kind='hist', color='orange')
plt.title("Top 10 products by Revenue")
plt.xlabel("Products")
plt.ylabel("Total Revenue")
plt.xticks(rotation=75)
plt.grid(True)
plt.tight_layout()
plt.show()
```
